```python
import numpy as np
```

```python
def activation(z):
  sigmoid = 1/(1+np.exp(-z))
  return sigmoid 
```

```python
input = np.array([[1,6,8,0],[33,2,3,5],[1,6,6,7],[2,9,8,4],[5,6,9,6],[32,13,0,2]])
output = np.array([[1,0,1,0],[0,1,0,1]])
print('Input :\n {},\n Output: \n{}'.format(input,output))
```

```python
weight = np.full((6,2),0.01).T
weight
```

```python
bias = np.full((2,1),0.01)
bias
```

```python
z = np.dot(weight , input) + bias
z
```

```python
y_predicted = activation(z)
y_predicted
```

```python
loss_mse = (1/6)*np.power(np.sum(output - y_predicted),2)
loss_mse
```
